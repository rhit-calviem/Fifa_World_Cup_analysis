```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from scipy.stats import linregress
from scipy.stats import ttest_rel
```

```python
df = pd.read_csv('final_result.csv')
df.head(3)
```

```python
scorer_counts = df['scorer'].value_counts()

# Select the top ten scorers
top_scorers = scorer_counts.head(10)

# Plot the bar chart
plt.figure(figsize=(10, 6))
top_scorers.plot(kind='bar')
plt.title('Top Ten Scorers')
plt.xlabel('Scorer')
plt.ylabel('Number of Goals')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# Filter out 'TIE' from game_winner column
df_filtered = df[df['game_winner'] != 'TIE']

# Count the occurrences of each team
teams = df_filtered['game_winner'].value_counts()

# Select the top ten teams
top_teams = teams.head(11)
# Plot the bar chart
plt.figure(figsize=(10, 6))
top_teams.plot(kind='bar')
plt.title('Top Teams')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
