```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from scipy.stats import linregress
from scipy.stats import ttest_rel
```

```python
df = pd.read_csv('results.csv')
df.head(3)
```

```python
scorer_counts = df['city'].value_counts()

# Select the top ten scorers
top_scorers = scorer_counts.head(10)

# Plot the bar chart
plt.figure(figsize=(10, 6))
top_scorers.plot(kind='bar')
plt.title('Top Cities')
plt.xlabel('City')
plt.ylabel('Games Hosted')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
scorer_counts = df['country'].value_counts()

# Select the top ten scorers
top_scorers = scorer_counts.head(10)

# Plot the bar chart
plt.figure(figsize=(10, 6))
top_scorers.plot(kind='bar')
plt.title('Top Countries')
plt.xlabel('Country')
plt.ylabel('Games Hosted')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
scorer_counts = df['tournament'].value_counts()

# Select the top ten scorers
top_scorers = scorer_counts.head(15)

# Plot the bar chart
plt.figure(figsize=(10, 6))
top_scorers.plot(kind='bar')
plt.title('Most played tournaments')
plt.xlabel('Tournament')
plt.ylabel('Number of games')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
